#### The leverage gained while modelling the data using decision tree is lost while an elsemble of such trees is made. It avarages out any information gain attained in smaller trees. In this case although accuracy increases the precision for 1 cases decrease.

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
print(nb.score(X_test,y_test))
print(confusion_matrix(nb.predict(X_test),y_test))
print(classification_report(nb.predict(X_test),y_test))
```

#### Clearly accuracy for attrition is a misleading parameter, one can say none of the employees fall in the group of attrition and still attain a respectable accuracy but does not provide any insight about which employee is going to leave the company next. On the other hand precision for 1 states is very important over recall for 1 states. The reason being the fact that those who have left the company nothing can be done about it; but information about those who might leave can be useful. If we focus on precision of 1 states, Gaussian NB model works comperatively better although accuracy is almost same to other models.
